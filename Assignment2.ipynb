
1. **Function to Check Number Type**
**Problem**

* Write a user-defined function that takes a number as input and returns:
 
  "Positive Even"
 
  "Positive Odd"
  
  "Negative"
  
  "Zero"

```python
def check_number(num):
    if num == 0:
        return "Zero"
    elif num > 0:
        if num % 2 == 0:
            return "Positive Even"
        else:
            return "Positive Odd"
    else:
        return "Negative"
```

* Explanation

  First, we check if the number is 0.
  
  If not zero, we check if it is positive.
  
  For positive numbers, % 2 helps us determine even or odd.
  
  If the number is neither zero nor positive, it must be negative.

2. **Function to Find the Largest of Three Numbers (Without max())**
**Problem**

* Write a function that accepts three numbers and returns the largest one without using built-in functions like max().


```python
def largest_of_three(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
```

* Explanation

  We compare a with b and c.
  
  If a is greater than or equal to both, it is the largest.
  
  Otherwise, we check b.
  
  If neither a nor b is largest, then c must be the largest.
  
  This conditional structure guarantees that only one value is returned as maximum.

3. **Function to Return Grade Based on Marks**
**Grading Logic**

   90+ → A
   
   75–89 → B
   
   50–74 → C
   
   Below 50 → Fail


```python
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 50:
        return "C"
    else:
        return "Fail"
```

* Explanation

  Conditions are checked from highest to lowest.
  
  Once a condition becomes true, remaining conditions are skipped.
  
  Nested decision-making ensures correct grade classification.
  
  This avoids overlapping conditions.

4. **Function to Validate Age**
**Problem**

   If age is negative → raise an error
   
   If age < 18 → "Minor"
   
   Otherwise → "Adult"


```python
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    elif age < 18:
        return "Minor"
    else:
        return "Adult"
```

* Explanation

  Input validation is done first by checking negative age.
  
  raise ValueError prevents invalid data from being processed.
  
  Remaining conditions classify the age properly.
  
  Conditional statements help ensure safe and valid input handling.


5. **Function to Process a List of Integers**
**Requirements**

   Count even numbers
   
   Ignore negative numbers
   
   Stop processing if 0 appears

```python
def count_even_numbers(lst):
    count = 0
    for num in lst:
        if num == 0:
            break
        elif num < 0:
            continue
        elif num % 2 == 0:
            count += 1
    return count
```

* Explanation

  A for loop is used to iterate through the list.
  
  break stops execution when 0 appears.
  
  continue skips negative numbers.
  
  Even numbers are counted using % 2 == 0.
  
  This function uses both loops and conditional statements effectively.
